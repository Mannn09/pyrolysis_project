```python
import numpy as np

class KineticsEngine:
    def __init__(self, config):
        self.cfg = config

    def update_mass(self, W, T, dt):
        T_K = T + 273.15
        for comp, (A_kin, Ea, n) in self.cfg.KINETICS_PARAMS.items():
            k_rate = A_kin * np.exp(-Ea / (self.cfg.R_GAS * T_K))
            W[comp] = np.maximum(W[comp] - (k_rate * (W[comp]**n) * dt), 0)
        return W
```
